```python
import pandas as pd
```

```python
h_list = ["landmark_id","category"]
df = pd.read_csv("train_label_to_category.csv",names=h_list)
```

```python
df = df[df['category'].str.contains("Tower|School|Building|Castle|University")]
```

```python
df
```

```python
df_split = df['category'].str.split(":",expand=True, n=2)
```

```python
df = df.drop(columns =['category'])
```

```python
df['landmark_name']=df_split[2]
```

```python
df['landmark_id'] = pd.to_numeric(df['landmark_id'])
```

```python
df.info()
```

```python
urls = pd.read_csv("train.csv")
```

```python
urls
```

```python
urls[urls["url"] != "None"]
```

```python
df = df.merge(urls, on='landmark_id')
#new_df = df[df["landmark_id"] == urls["landmark_id"]]
```

```python
df = df.drop_duplicates(subset=['landmark_id'], keep='last')
```

```python
df
```

```python

from PIL import Image
import urllib
import requests
from io import BytesIO

for index, row in df.iterrows():
    imageurl = str(row['url'])
    print(imageurl)
    try:
        img = Image.open(urllib.request.urlopen(imageurl))
        img.save(str(row['landmark_name'])+".jpg")
    except:
        print("Oops!  That was no valid number.  Try again...")
```
